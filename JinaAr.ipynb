```python
def jaccard(data):
    if data.shape[0] <2 :
        return
    
    a=data.iloc[0]['Text'].split(',')
    b=data.iloc[1]['Text'].split(',')
    union = list(set(a+b))
    intersection = list(set(a) - (set(a)-set(b)))
    #print ("Union - " , union)
    #print ("Intersection - ", intersection)
    jaccard_coeff = float(len(intersection))/len(union)
    #print ("Jaccard Coefficient is =  " , jaccard_coeff)
      
    if data.shape[0] >2 :
        #print ("3333333")
        c=data.iloc[2]['Text'].split(',')
        union = list(set(a+c))
        intersection = list(set(a) - (set(a)-set(c)))
        jaccard_coeff2 = float(len(intersection))/len(union)
        jaccard_coeff = (jaccard_coeff + jaccard_coeff2)/2
        
    return data, jaccard_coeff
```

```python
df3 = df2.groupby(['Teacher','ClassName','Group','Final']).apply(jaccard)
df3.reset_index()
df3.to_csv('jaccard.csv')
```
